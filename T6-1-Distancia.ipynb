# Distancias

```python
import pandas as pd
from scipy.spatial import distance_matrix
```

```python
data=pd.read_csv("../datasets/movies/movies.csv",sep=";")
data
```

```python
movies = data.columns.values.tolist()[1:] #saco el id
movies
```

**Aqui no hace falta normalizar, pq todos los valores van de 0 a  10**

```python
dd1 = distance_matrix(data[movies],data[movies],p=1) #distancia de manhatan
dd2= distance_matrix(data[movies],data[movies],p=2) #distancia euclidea
dd10 = distance_matrix(data[movies],data[movies],p=10)
```

```python
def dm_to_df(dd,col_name):
    import pandas as pd
    return pd.DataFrame(dd,index=col_name,columns=col_name)
```

```python
dm_to_df(dd1,data["user_id"])
```

```python
dm_to_df(dd2,data["user_id"])
```

```python
dm_to_df(dd10,data["user_id"])
```

## Clustering jerárquico 

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

**Con parámetro "ward" realiza el enlace con distancias de cuadrados, tenemos varios**

```python
z = linkage(data[movies], "ward")  ## une los eleemntos x, y a una distancia de d y te dice cuantos 
z                                    ##elementos habia en ese grupo
```

**Ahora podemos pintar un dendrograma**

```python
plt.figure(figsize=(25,10))##tamaño grande
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Usuarios de Netflix")
plt.ylabel("distancia")
dendrogram(z,leaf_rotation=90.,leaf_font_size=10)
plt.show()
```

**podemos ponerlo en horizontal**

```python
z = linkage(data[movies], "average")  ## distancia promedio 
plt.figure(figsize=(25,10))           ##tamaño grande
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Usuarios de Netflix")
plt.ylabel("distancia")
dendrogram(z,leaf_rotation=90.,leaf_font_size=10, orientation="right")
plt.show()
```

```python
data[movies]
```

```python
z = linkage(data[movies], "complete")  ## enlace completo
plt.figure(figsize=(25,10))           ##tamaño grande
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Usuarios de Netflix")
plt.ylabel("distancia")
dendrogram(z,leaf_rotation=90.,leaf_font_size=10)
plt.show()
```

```python
z = linkage(data[movies], "single")  ## enlace completo
plt.figure(figsize=(25,10))           ##tamaño grande
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Usuarios de Netflix")
plt.ylabel("distancia")
dendrogram(z,leaf_rotation=90.,leaf_font_size=10)
plt.show()
```

**Para mas información de los diferentes links, ver en linkage python. Con el parámetro de métrica, se pueden indicar las distancias de manhatan, coseno, minkowski etc. La del coseno se ocupa para valores binarios generalemente**
