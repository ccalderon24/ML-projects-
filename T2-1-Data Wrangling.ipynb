# Data Wrangling (cirugía de datos)

EL data wrangling a veces denominada data munging, es el proceso de transformar y mapear datos de un dataset raw (en bruto) en otro formato con la intención de hacerlo más apropiado y valioso para una variedad de propósitos posteriores, como el análisis. Un data wrangler es una persona que realiza estas operaciones de transformación.

Esto puede incluir munging, visualización de datos, agregación de datos, entrenamiento de un modelo estadístico, así como muchos otros usos potenciales. La oscilacion de datos como proceso generalmente sigue un conjunto de pasos generales que comienzan extrayendo los datos de forma cruda de origen de datos, dividiendo los datos en bruto usando algoritmos(por ej: clasificacion) o analizando los datos en estructuras de datos predefinidas, y finalmente depositando el conetnido resultando en un sistema de almacenamiento para su uso futuro

## Generación aleatoria de números

```python
import numpy as np
```

```python
# generar un numero aleatorio entero entre 1 y 100
np.random.randint(1,100)
```

```python
# la forma clasica es generar un numero aleatorio entre 0 y 1 (con decimales)
np.random.random()
```

```python
##fncion que genera una lista de n numeros aleatorios enteros dentro del interbalo [a,b]
def randint_list(n,a,b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(25,1,50)
```

## seed

cuando se necesite retener una cantidad de numeros aleatorios que te gustaron es necesario fijar la semilla

```python
np.random.seed(2018)
for i in range(5):
    print(np.random.random())
```

es importante que cuando tienes un proyecto y quieres realizar numeros aleatorios al azar, fijar la semilla arriba del programa, de esa forma te aseguras de que podrás reproducir tus valores cuando desees
