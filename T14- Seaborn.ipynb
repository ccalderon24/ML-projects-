# Gráficos de cuadrículas (grids): PairGrid y FacetGrid

```python
flores = sns.load_dataset("iris")
flores.head()
```

## PairGrid

```python
sns.pairplot(flores)
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter)
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot)  ## solo diagonal
graficos.map_upper(plt.scatter)  ## solo parte superior de la diag (lower para la inf)
graficos.map_lower(sns.kdeplot)  ## inferior
```

```python
g = sns.pairplot(flores, corner=True, diag_kind= "kde")
```
