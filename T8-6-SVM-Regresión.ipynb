## SVM para Regresi√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.sort(5*np.random.rand(200,1),axis=0)
y = np.sin(x).ravel()

y[::5]+= 3*(0.5 - np.random.rand(40))   # aqui introducimos ruido para el problemas
```

```python
plt.scatter(x,y, color="darkorange",label="data")
```

```python
from sklearn.svm import SVR
```

```python
c=1e3
svr_lin = SVR(kernel="linear",C=c)
svr_rbf = SVR(kernel="rbf",C=c, gamma=0.1)
svr_pol = SVR(kernel="poly",C=c, degree=3)
```

```python
y_lin = svr_lin.fit(x,y).predict(x)
y_rbf = svr_rbf.fit(x,y).predict(x)
y_pol = svr_pol.fit(x,y).predict(x)
```

```python
lw = 2
plt.figure(figsize= (16,9))
plt.scatter(x,y,color="darkorange",label="data")
plt.plot(x,y_lin,color="navy", lw=lw, label="SVM lineal")
plt.plot(x,y_rbf,color="c", lw=lw, label="SVM RBF")
plt.plot(x,y_pol,color="cornflowerblue", lw=lw, label="SVM polinomico")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vector Regression")
plt.legend()
plt.show()
```
